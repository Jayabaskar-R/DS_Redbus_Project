```python
import pandas as pd 

df1 = pd.read_csv("ap_bus_details.csv")
df2 = pd.read_csv("assam_bus_details.csv")
df3 = pd.read_csv("chandigarh_bus_details.csv")
df4 = pd.read_csv("himachal_bus_details.csv")
df5 = pd.read_csv("kerala_bus_details.csv")
df6 = pd.read_csv("rajasthan_bus_details.csv")
df7 = pd.read_csv("Telangana_bus_details.csv")
df8 = pd.read_csv("up_bus_details.csv")
df9 = pd.read_csv("wb_bus_details.csv")
df10 = pd.read_csv("jk_bus_details.csv")

csv_files = ["ap_bus_details.csv", "assam_bus_details.csv", "chandigarh_bus_details.csv", "himachal_bus_details.csv", "kerala_bus_details.csv","rajasthan_bus_details.csv", "Telangana_bus_details.csv", "up_bus_details.csv", "wb_bus_details.csv",  "jk_bus_details.csv"]   

# Read each CSV file into a DataFrame and store it in a list
df = [pd.read_csv(file) for file in csv_files]

# Concatenate all DataFrames in the list
combined_df = pd.concat(df, ignore_index=True)

combined_df.to_csv("bus_routes.csv", index=False)


for i in range(len(combined_df)):
    mycursor.execute("""INSERT INTO  bus_details(Route_name, Route_link, Bus_name, Bus_type, Departing_time, Duration, Reaching_time, Star_rating, Price, Seat_availability)    
    values (%s, %s, %s, %s, %s, %s, %s, %s, %s,%s)""", [combined_df["Route_Name"].loc[i], combined_df["Route_Link"].loc[i], combined_df["Bus_Name"].loc[i], combined_df["Bus_Type"].loc[i], combined_df["Departing_Time"].loc[i], combined_df["Duration"].loc[i], combined_df["Reaching_Time"].loc[i], float(combined_df["Star_Rating"].loc[i]), combined_df["Price"].loc[i], combined_df["Seat_Availability"].loc[i]])
    mydb.commit()
mycursor.close()

```
